**1. El modelo de Lotka-Volterra (predador-presa)**

Inciso a)

```python
import numpy as np
import matplotlib.pyplot as plt

#parámetros 
a = 1
b = 0.5
g = 0.5
d = 2

#condiciones iniciales
x0 = 2
y0 = 2

#tiempos
t0 = 0
tf = 10
dt = 0.01
n = int((tf-t0)/dt)

#definir los arreglos
t = np.linspace(t0, tf, n)
x = np.zeros(n)
y = np.zeros(n)
x[0] = x0
y[0] = y0

#método de euler
for i in range(1,n):
    x[i] = x[i-1] + dt*(a*x[i-1] - b*x[i-1]*y[i-1])
    y[i] = y[i-1] + dt* (g*x[i-1]*y[i-1] - d*y[i-1])

plt.plot(t, x, label = "Conejos(x)")
plt.plot(t, y, label = "Zorros (y)")
plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.legend()
plt.title("Modelo de Lotka-Volterra por método de euler")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#parámetros 
a = 1
b = 0.5
g = 0.5
d = 2

#condiciones iniciales
x0 = 2
y0 = 2

#tiempos
t0 = 0
tf = 10
dt = 0.01
n = int((tf-t0)/dt)

#definir los arreglos
t = np.linspace(t0, tf, n)
x = np.zeros(n)
y = np.zeros(n)
x[0] = x0
y[0] = y0

#definir derivadas
def dx(x,y):
    return((a*x)-(b*x*y))

def dy(x,y):
    return((g*x*y)-(d*y))

#método de runge kutta
for i in range(1,n):
    k1x = dt*dx(x[i-1], y[i-1])
    k1y = dt*dy(x[i-1], y[i-1])
    k2x = dt*dx(x[i-1]+(0.5*k1x), y[i-1]+(0.5*k1y))
    k2y = dt*dy(x[i-1]+(0.5*k1x), y[i-1]+(0.5*k1y))
    k3x = dt*dx(x[i-1]+(0.5*k2x), y[i-1]+(0.5*k2y))
    k3y = dt*dy(x[i-1]+(0.5*k2x), y[i-1]+(0.5*k2y))
    k4x = dt*dx(x[i-1]+k3x, y[i-1]+ k3y)
    k4y = dt*dy(x[i-1]+k3x, y[i-1]+ k3y)
    x[i] = x[i-1]+(k1x+(2*k2x)+(2*k3x)+k4x)/6
    y[i] = y[i-1]+(k1y+(2*k2y)+(2*k3y)+k4y)/6

plt.plot(t, x, label = "Conejos(x)")
plt.plot(t, y, label = "Zorros (y)")
plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.legend()
plt.title("Modelo de Lotka-Volterra por método de Runge-Kutta de cuarto orden")
plt.grid()
plt.show()
```

Para la energía cinética

La velocidad de la primera masa es

$$v_1=\ell\dot{\theta_1}$$ 
La velocidad de la segunda mas es

$$v_2=\sqrt{(\ell\dot{\theta_1}\cos\theta_1+\ell\dot{\theta_2}\cos\theta_2)^2+(\ell\dot{\theta_1}\sin\theta_1+\ell\dot{\theta_2}\sin\theta_2)^2}$$

Podemos obtener dos energías cinéticas:

$$T_1=\frac{1}{2}m(\ell\dot{\theta_1})^2=\frac{1}{2}m\ell^2\dot{\theta_1}^2$$ 

$$T_2=\frac{1}{2}m\ell^2(\dot{\theta_1}^2+\dot{\theta_2}^2+2\dot{\theta_1}\dot{\theta_2}\cos(\theta_1-\theta_2))$$ 
Obteniendo la energía cinética total sumando las $T_1$ y $T_2$ obtenemos por lo tanto que

$$T=\frac{1}{2}m\ell^2(2\dot{\theta_1}^2+\dot{\theta_2}^2+2\dot{\theta_1}\dot{\theta_2}\cos(\theta_1-\theta_2))$$

Para conocer la energía potencial, podemos definir la altura para cada masa como

$$h_1=-\ell\cos\theta_1$$
Y para la masa segunda masa como
$$h_2=-\ell(\cos\theta_1+\cos\theta_2)$$ 

Entonces para cada energía potencial encontramos que:
$$\nu_1=-mg\ell\cos\theta_1$$ 

$$\nu_2=-mg\ell(\cos\theta_1+\cos\theta_2)$$ 
Para la energía potencial total obtenemos que:

$$\nu=-mg\ell(2\cos\theta_1+\cos\theta_2)$$ 

Por lo tanto sumando $T$ y $\nu$ obtenemos que la energía total es:

$$E=\frac{1}{2}m\ell^2(2\dot{\theta_1}^2+\dot{\theta_2}^2+2\dot{\theta_1}\dot{\theta_2}\cos(\theta_1-\theta_2))-mg\ell(2\cos\theta_1+\cos\theta_2)$$

**3. Ecuacion de Poisson para un potencial electrostático**

```python
import numpy as np
import matplotlib.pyplot as plt

L = 1 #largo de la caja en metros
N = 100 #numero de puntos en la cuadricula
dx = L/(N-1) #espacios en la cuadricula
e0 = 1 #permitividad como lo pide el problema
phi = np.zeros((N,N)) #matriz de ceros para la cuadricula de potencial y densidad de carga
rho = np.zeros((N,N))
#definir las posciones y tamaño de la carga
tcarga = int(0.2 / dx) #Tamaño de la carga en puntos de la cuadricula
des = int(0.2/dx) #desplazamiento desde las paredes en los puntos de la cuadricula
#carga positiva
rho[des:des+tcarga, des:des+tcarga] = 1
#carga negativa 
rho[-(des+tcarga):-des,-(des+tcarga):-des] = -1
#metodo de relajacion
tol = 1e-6 #tolerancia
maxit = 10000 #iteraciones máximas
for it in range(maxit):
    phio = phi.copy()
    phi[1:-1, 1:-1] = 0.25 * (phio[1:-1, :-2] + phio[1:-1, 2:] + phio[:-2, 1:-1] + phio[2: , 1:-1] - dx**2 * rho[1:-1, 1:-1]/e0)
    #condiciones de frontera
    phi[0, :] = phi[-1, :] = phi[:,0] = phi[:, -1] = 0
    #verificación de convergencia
    if np.max(np.abs(phi-phio))<tol:
        print("Convergencia alcanzada en", it, "iteraciones")
        break
    if it == maxit-1:
        print("no se alcanzó la convergencia")

#obtener el potencial en un punto dado
def obphi(x,y):
    i = int(x/dx)
    j = int(y/dx)
    return(phi[i,j])

#ejemplo 
xp = 0.5
yp = 0.5
potencial = obphi(xp, yp)
print("El potencial electrostático en el punto", (xp, yp), "es", potencial)
```
